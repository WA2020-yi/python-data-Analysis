# **项目：评估和清理英国电商公司销售数据**

### **分析目标**

此数据分析的目的是，根据市场销售数据，挖掘畅销产品，以便制定更有效的市场策略来提升营收。

本实战项目的目的在于练习评估数据干净和整洁度，并且基于评估结果，对数据进行清洗，从而得到可供下一步分析的数据。


### **简介**

原始数据集记录了一家英国在线零售公司在2010年12月1日至2011年12月9日期间的所有交易情况，涵盖了该公司在全球不同国家和地区的业务数据。该公司主要销售覆盖各个场景的礼品，包括但不限于生日礼品、结婚纪念品、圣诞礼品等等。该公司的客户群体主要包括批发商和个人消费者，其中批发商占据了相当大的比例。


数据每列含义如下：

- InvoiceNo：发票号码。6位数，作为交易的唯一标识符。如果这个代码以字母"c"开头，表示这笔交易被取消。
- StockCode：产品代码。5位数，作为产品的唯一标识符。
- Description：产品名称。
- Quantity：产品在交易中的数量。
- InvoiceDate：发票日期和时间。交易发生的日期和时间。
- UnitPrice：单价。价格单位为英镑（£）。
- CustomerID：客户编号。5位数，作为客户的唯一标识符。
- Country：国家名称。客户所居住的国家的名称。





## **读取数据**

```python
import pandas as pd
```

```python
original_data = pd.read_csv("D:/data.csv",encoding="ISO-8859-1")
```

```python
original_data.head()
```

## **评估数据**

评估数据从结构和内容两方面进行，即整洁度和干净度。数据结构性问题指不符合"每个变量为一列，每个观察值为一行，每个单元格为一个值"这三个标准，数据内容性包括存在丢失数据、重复数据、不一致数据、错误或无效数据等

```python
original_data.sample(10)
```

从以上观察来看，数据符合每列是一个变量，每行是一个观察值，每个单元格是一个值的标准，结构性没有问题。

### 评估数据干净度

```python
original_data.info()
```

通过以上观察，发现"Description"和"CustomerID"这两个变量存在缺失值，  
另外，InvoiceDate这个变量指的是发票日期和时间，类型应为日期，CustomerID指客户编号，类型应为字符串

### 评估缺失数据

在了解Description存在缺失值后，根据条件提取出缺失观察值

```python
original_data[original_data["Description"].isnull()]
```

通过以上观察，发现Description为空时，UnitPrice都为0，我们需要确认是否有Description变量缺失且UnitPrice不为零的观察值

```python
original_data[(original_data["Description"].isnull()) & (original_data["UnitPrice"] != 0)]
```

可以看到输出结果是空的，说明没有任何的观察值是Description为空且UnitPrice为0，  
也就是说缺失Description的交易数据，也不具备有效UnitPrice值

Description表示产品名称，UnitPrice表示产品单价，都是进行后续商品交易分析的重要变量，若是它们同时缺失/无效，我们认为数据无法提供有效含义，因此这些后续可以删除

CustomerID变量同样存在缺失值，因此也根据条件提取出缺失观察值

```python
original_data[original_data["CustomerID"].isnull()]
```

CustomerID表示客户编号，不是分析畅销产品的必要变量。从输出结果看，有些缺失CustomerID的交易数据仍然有效，因为保留此变量为空的观察值

### 评估重复数据

根据变量含义来看，InvoiceNo(发票号码)，StockCode(产品代码)，CustomerID(客户编号)都是唯一标识符，但一次交易可能包含多件商品，所以InvoiceNo可以重复，不同交易可以包含同一件商品，所以StockCode可以重复，客户可以多次下单，所以CustomerID可以重复。
那么针对此数据集，我们无需评估重复数据

### 评估不一致数据

不一致数据可能存在于"country"变量中，我们要查看是否存在多个不同值指代同一个国家

```python
original_data["Country"].value_counts()
```

从Country变量值来看，并不存在不一致数据

### 评估无效/错误数据

可以通过DataFrame的describe方法，对数值统计信息进行快速了解

```python
original_data.describe()
```

发现Quantity(产品交易数量)和UnitPrice(单价)变量存在负数，会对后续数值分析产生影响。
因此，我们先筛选出Quantity数值为负数的观察值，进一步评估其含义

```python
original_data[original_data["Quantity"] < 0]
```

从筛选结果看，Quantity变量为负数的观察值，InvoiceNo似乎都是以C开头，表示订单被取消

为了进一步确认，我们增加筛选条件，看是否存在Quantity为负数，InvoiceNo不以C开头的观察值

```python
original_data[(original_data["Quantity"] < 0) & (original_data["InvoiceNo"].str[0] != "C")]
```

观察结果发现，Quantity为负数与被取消订单并无关联，但同时我们又发现当Quantity为负数，InvoiceNo不以C开头的观察值，产品单价似乎都为0

```python
original_data[(original_data["Quantity"] < 0) & (original_data["InvoiceNo"].str[0] != "C") & (original_data["UnitPrice"] != 0)]
```

根据输出结果，发现当Quantity变量为负数的时候，观察值满足一下条件之一：

InvoiceNo 以C开头，表示订单被取消  
UnitPrice 为0，说明单价为0英镑

这些交易数据都不是有效成交数据，不贡献销售，不在后续分析范围内，因此将在清理数据步骤，将Quantity为负数的观察值删除

接下来筛选出UnitPrice为负数的观察值

```python
original_data[original_data["UnitPrice"] < 0]
```

可以看到Description下的观察值为Adjust bad debt(坏账调整)，不属于实际商品交易数据，将在清理数据步骤中删除

## **清理数据**

根据前面评估部分得到的结论，我们需要进行清理的数据包括：  
把 InvoiceDate 变量的数据类型转为日期  
把 CustomerID 变量的数据类型转为字符串  
把 Description 变量缺失的观察值删除  
把 Quantity 变量值为负数的观察值删除  
把 UnitPrice 变量值为负数的观察值删除

为了区分开经过清理后的数据和原始的数据，我们创建新的变量cleaned_data，让它为original_data复制出的副本。我们之后的清理步骤都将被运用在cleaned_data上

```python
cleaned_data = original_data.copy()
cleaned_data.head()
```

把 InvoiceDate 变量的数据类型转换为日期时间类型

```python
cleaned_data["InvoiceDate"] = pd.to_datetime(cleaned_data["InvoiceDate"])
cleaned_data["InvoiceDate"]
```

把 CustomerID 变量的数据类型转换为字符串：

```python
cleaned_data["CustomerID"] = cleaned_data["CustomerID"].astype(str)
cleaned_data["CustomerID"]
```

把 CustomerID 变量结尾的 ".0" 删除：

```python
cleaned_data["CustomerID"] = cleaned_data["CustomerID"].str.slice(0,-2)
cleaned_data["CustomerID"]
```

把 Description 变量缺失的观察值删除,并查看删除后该列空缺值个数和：

```python
cleaned_data.dropna(subset = ["Description"],inplace=True)
```

```python
cleaned_data["Description"].isnull().sum()
```

把 Quantity 变量值为负数的观察值删除，并检查替换后Quantity变量值为负数的个数：

```python
cleaned_data = cleaned_data[cleaned_data["Quantity"] >= 0]
```

```python
len(cleaned_data[cleaned_data["Quantity"] < 0])
```

把 UnitPrice 变量值为负数的观察值删除，并检查替换后UnitPrice变量值为负数的个数：

```python
cleaned_data = cleaned_data[cleaned_data["UnitPrice"] >= 0]
```

```python
len(cleaned_data[cleaned_data["UnitPrice"] < 0])
```

## **保存清理后的数据**

完成数据清理后，把干净整洁的数据保存到新的文件里，文件名为data_cleaned.csv
